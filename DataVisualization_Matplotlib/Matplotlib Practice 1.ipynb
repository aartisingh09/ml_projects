```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL 
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
#print('string')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
#OBJECT ORIENTED METHOD
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.7])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fg = plt.figure()

axes1 = fg.add_axes([0.1,0.1,0.8,0.8])
axes2 = fg.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)

axes2.set_title('SMALLER PLOT')
```
