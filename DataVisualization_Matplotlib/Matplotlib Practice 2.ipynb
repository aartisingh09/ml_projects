```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.7])
axes1.plot(x,y)
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
```

```python
fig,axes1 = plt.subplots(nrows=1, ncols=2)

for current_ax in axes1:
    current_ax.plot(x,y)
```

```python
axes1
```

```python
fig,axes1 = plt.subplots(nrows=1, ncols=2)

axes1[0].plot(x,y)
axes1[0].set_title('frst')
axes1[0].set_xlabel('bla')

axes1[1].plot(y,x)
axes1[1].set_title('second')

plt.tight_layout()
```

Fgure Size and DPI

```python
fig,axes = plt.subplots(nrows=2 , ncols= 1, figsize=(3,2), dpi = 100)

#ax = fig.add_axes([0,0,1,1])
axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')

ax.plot(x,x**3, label= 'X cubed')

ax.legend()
```
